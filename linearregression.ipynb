```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year,df.income)
```

```python
year = df.drop('income',axis='columns')
year
```

```python
income = df.income
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```
